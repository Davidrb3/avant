```python
from fuzzywuzzy import fuzz
from difflib import SequenceMatcher
```

```python
def fuzzywuzzy_matching(data1, data2, truncate, threshold, flag):
    minratio_list = []
    volume_author_list = []
    review_author_list = []
    aRatio_list = []
    volume_title_list = []
    review_title_list = []
    tRatio_list = []
    volume_year_list = []
    review_year_list = []
    for i in range(len(data1)):
        if i%10000 == 0:
            print(i)
        author = data1['author'][i]
        if author != author: # do not count the case when author is nan
            continue
        author_initial = author[0]
        title = data1['title'][i]
        year = data1['latestcomp'][i]
        for j in range(len(data2)):
            author2 = data2['author'][j]
            if author2 != author2: # do not count the case when author is nan
                continue
            author2_initial = author2[0]
            if author_initial != author2_initial: # blocking: only match the case where the authors' first initials match
                continue
            else:
                title2 = data2['title'][j]
                if truncate == 0:
                    new_title = title
                    new_title2 = title2
                else:
                    if flag == 1:
                        shorter_title = min(max(len(title), len(title2)), truncate) # determining the truncate length: shorter one of the longer title and truncate number
                    else:
                        shorter_title = min(len(title), len(title2), truncate) # determining the truncate length: shortest of two titles and truncate number
                    new_title = title[:shorter_title]
                    new_title2 = title2[:shorter_title]
                year2 = data2['year'][j]
                aRatio = fuzz.ratio(str(author).lower(), str(author2).lower())
                tRatio = fuzz.ratio(str(new_title).lower(), str(new_title2).lower())
                minratio = min(aRatio, tRatio)
                if minratio >= threshold:
                    volume_author_list.append(author)
                    review_author_list.append(author2)
                    aRatio_list.append(aRatio)
                    volume_title_list.append(new_title)
                    review_title_list.append(new_title2)
                    tRatio_list.append(tRatio)
                    volume_year_list.append(year)
                    review_year_list.append(year2)
                    minratio_list.append(minratio)
    df_output = pd.DataFrame(list(zip(minratio_list, volume_author_list, review_author_list,  aRatio_list, volume_title_list, review_title_list,
                                  tRatio_list, volume_year_list, review_year_list)), columns=['minRatio', 'volume_author', 'review_author',
                                                                                              'aRatio', 'volume_title', 'review_title',
                                                                                              'tRatio', 'volume_year', 'review_year'])
    return df_output
```

```python
# processes about 500 records in 20s. Takes less than 2h to finish
fuzzywuzzy_full = fuzzywuzzy_matching(volume_title_author_year, review_title_author_year, 0, 60, 0)
fuzzywuzzy_full = fuzzywuzzy_full.sort_values(by=['minRatio'], ascending=False).reset_index(drop=True)
fuzzywuzzy_full
```

```python
def sequence_matching(data1, data2, truncate, threshold, flag):
    minratio_list = []
    volume_author_list = []
    review_author_list = []
    aRatio_list = []
    volume_title_list = []
    review_title_list = []
    tRatio_list = []
    volume_year_list = []
    review_year_list = []
    for i in range(len(data1)):
        if i%10000 == 0:
            print(i)
        author = data1['author'][i]
        if author != author: # do not count the case when author is nan
            continue
        author_initial = author[0]
        title = data1['title'][i]
        year = data1['latestcomp'][i]
        for j in range(len(data2)):
            author2 = data2['author'][j]
            if author2 != author2: # do not count the case when author is nan
                continue
            author2_initial = author2[0]
            if author_initial != author2_initial: # blocking: only match the case where the authors' first initials match
                continue
            else:
                title2 = data2['title'][j]
                if truncate == 0:
                    new_title = title
                    new_title2 = title2
                else:
                    if flag == 1:
                        shorter_title = min(max(len(title), len(title2)), truncate) # determining the truncate length: shorter one of the longer title and truncate number
                    else:
                        shorter_title = min(len(title), len(title2), truncate) # determining the truncate length: shortest of two titles and truncate number
                    new_title = title[:shorter_title]
                    new_title2 = title2[:shorter_title]
                year2 = data2['year'][j]
                aRatio = SequenceMatcher(None, str(author).lower(), str(author2).lower()).ratio()
                tRatio = SequenceMatcher(None, str(new_title).lower(), str(new_title2).lower()).ratio()
                minratio = min(aRatio, tRatio)
                if minratio >= threshold:
                    volume_author_list.append(author)
                    review_author_list.append(author2)
                    aRatio_list.append(aRatio)
                    volume_title_list.append(new_title)
                    review_title_list.append(new_title2)
                    tRatio_list.append(tRatio)
                    volume_year_list.append(year)
                    review_year_list.append(year2)
                    minratio_list.append(minratio)
    df_output = pd.DataFrame(list(zip(minratio_list, volume_author_list, review_author_list,  aRatio_list, volume_title_list, review_title_list,
                                  tRatio_list, volume_year_list, review_year_list)), columns=['minRatio', 'volume_author', 'review_author',
                                                                                              'aRatio', 'volume_title', 'review_title',
                                                                                              'tRatio', 'volume_year', 'review_year'])
    return df_output
```
