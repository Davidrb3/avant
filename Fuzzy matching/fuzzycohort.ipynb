```python
def fuzzywuzzy_matching(data1, data2, truncate, auththreshold, titlethreshold):

    volume_author_list = []
    review_author_list = []
    aRatio_list = []
    volume_title_list = []
    review_title_list = []
    truncated_rev_title_list = []
    tRatio_list = []
    volume_year_list = []
    review_year_list = []
    docid_list = []
    rownum_list = []
    
    data1['author'] = data1.author.str.strip('"[')
    data2['author'] = data2.author.str.strip('"[')
    
    data1['initial'] = data1['author'].str[0]
    data2['initial'] = data2['author'].str[0]
    
    allinitials = set(data1.initial).intersection(set(data2.initial))
    matches = 0
    
    for initial in allinitials:  
        print(initial)
        df1 = data1.loc[data1.initial == initial, : ]
        df2 = data2.loc[data2.initial == initial, : ]
        print(len(df1['author']), len(df2))
        
        for idx, row in df1.iterrows():
            author = row.author
            if author != author: # do not count the case when author is nan
                continue
            title = row.title
            if pd.isnull(title):
                    continue
                    
            year = row.year
            
            idxlist = []
            
            for idx2, row2 in df2.iterrows():
                author2 = row2.author
                if author2 != author2: # do not count the case when author is nan
                    continue
               
                aRatio = fuzz.ratio(str(author), str(author2))
                if aRatio < auththreshold:
                    continue
                
                title2 = row2.title
                if pd.isnull(title2):
                    continue
                
                if title.startswith('the '):        # remove initial the before truncation
                    title = title[4:]               # this is commonly missing in the reviews
                if title2.startswith('the '):
                    title2 = title2[4:]
                
                if truncate == 0:
                    new_title = title
                    new_title2 = title2
                    
                else:       
                    shorter_title = min(len(title), len(title2), truncate) # determining the truncate length: shortest of two titles and truncate number
                    new_title = title[:shorter_title]
                    new_title2 = title2[:shorter_title]
                
                minlen = min(len(new_title), len(new_title2))
                
                if minlen < 3:
                    continue
                
                if minlen < 12:
                    handicap = 12 - minlen
                elif minlen > 24:
                    handicap = 24 - minlen  # this will actually give an advantage to long titles, up to truncate max
                else:
                    handicap = 0
                
                    
                year2 = row2.year
                    
                tRatio = fuzz.ratio(str(new_title), str(new_title2)) - handicap
                
                totalmatch = tRatio + (aRatio / 1.5)
                
                if tRatio >= titlethreshold:
                    idxlist.append((totalmatch, idx2, aRatio, tRatio, new_title2, year2, row2.rownum))
            
            if len(idxlist) > 0:
                idxlist.sort(reverse = True)
                row2 = df2.loc[idxlist[0][1], : ]
                aRatio = idxlist[0][2]
                tRatio = idxlist[0][3]
                new_title2 = idxlist[0][4]
                year2 = idxlist[0][5]
                rownum = idxlist[0][6]
                
                volume_author_list.append(author)
                author2 = row2.author
                review_author_list.append(author2)
                
                aRatio_list.append(aRatio)
                volume_title_list.append(new_title)
                old_title = row2.title
                review_title_list.append(old_title)
                truncated_rev_title_list.append(new_title2)
                
                tRatio_list.append(tRatio)
                volume_year_list.append(year)
                review_year_list.append(year2)
                docid_list.append(row.docid)
                rownum_list.append(rownum)
                
                matches += 1
                if matches % 100 == 1:
                    print('matches: ', matches)
                    
    df_output = pd.DataFrame(list(zip(volume_author_list, review_author_list,  aRatio_list, volume_title_list, truncated_rev_title_list, review_title_list,
                                  tRatio_list, volume_year_list, review_year_list, docid_list, rownum_list)), columns=['volume_author', 'review_author',
                                                                                              'aRatio', 'volume_title', 'trunc_revtitle', 'review_title',
                                                                                              'tRatio', 'volume_year', 'review_year', 'docid', 'rownum'])
    return df_output
```

```python
# processes about 500 records in 20s. Takes less than 2h to finish
fuzzywuzzy_full = fuzzywuzzy_matching(vtay, review_title_author_year, 30, 70, 80)
fuzzywuzzy_full = fuzzywuzzy_full.sort_values(by=['tRatio'], ascending=False).reset_index(drop=True)
fuzzywuzzy_full.head()
```
