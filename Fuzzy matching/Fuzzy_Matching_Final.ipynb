```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 200)
```

```python
titlemeta = pd.read_csv("D:/DH collaborative/Avant garde/titlemeta.tsv", sep="\t", low_memory=False)
titlemeta
```

```python
titlemeta_clean = titlemeta.copy(deep = True)
titlemeta_clean = titlemeta_clean[['docid', 'title', 'author', 'latestcomp', 'genres', 'subjects', 'authordate', 'imprint', 'enumcron']]
titlemeta_clean['title'] = titlemeta['title'].str.replace('\| ', '', regex = True)
titlemeta_clean['title'] = titlemeta_clean['title'].replace('\/ ', '', regex = True)
titlemeta_clean['title'] = titlemeta_clean['title'].replace('\$c', '', regex = True)
titlemeta_clean['title'] = titlemeta_clean['title'].replace(': ', '', regex = True)
titlemeta_clean
```

```python
volume_title_author_year = titlemeta_clean[['title', 'author', 'latestcomp', 'docid']]
volume_title_author_year = volume_title_author_year.rename(columns = {'title': 'title', 'author': 'author', 'latestcomp': 'year', 'docid' : 'docid'})
volume_title_author_year['title'] = volume_title_author_year['title'].str.lower()
volume_title_author_year['author'] = volume_title_author_year['author'].str.lower()
volume_title_author_year
```

```python
revs = list(range(10))
for i in range(9):
    link = "D:/DH collaborative/Avant garde/1930srevs/nonconsumptive" + str(i +27) + "_193" + str(i + 1) + ".tsv"
    revs[i] = pd.read_csv(link, sep="\t", low_memory=False)
    revs[i]['year'] = 1931 + i
    revs[i]['rownum'] = [x for x in range(len(revs[i]))]
revs[9] = pd.read_csv("D:/DH collaborative/Avant garde/1930srevs/nonconsumptive36_1940.tsv", sep="\t", low_memory=False)
revs[9]['year'] = 1940
revs[9]['rownum'] = [x for x in range(len(revs[9]))]
review_meta = pd.concat(revs, axis=0).reset_index(drop=True)
review_meta
```

```python
review_title_author_year = review_meta[['booktitle', 'bookauthor', 'year', 'rownum']].reset_index(drop=True)
review_title_author_year.columns = ['title', 'author', 'year', 'rownum']
review_title_author_year['title'] = review_title_author_year['title'].str.lower()
review_title_author_year['author'] = review_title_author_year['author'].str.lower()
review_title_author_year
```

```python
from fuzzywuzzy import fuzz
def fuzzywuzzy_matching(data1, data2, truncate, auththreshold, titlethreshold):

    volume_author_list = []
    review_author_list = []
    aRatio_list = []
    volume_title_list = []
    truncated_volume_title_list = []
    review_title_list = []
    truncated_rev_title_list = []
    tRatio_list = []
    original_tRatio_list = []
    volume_year_list = []
    review_year_list = []
    docid_list = []
    rownum_list = []
    
    data1['author'] = data1.author.str.strip('"[')
    data2['author'] = data2.author.str.strip('"[')
    
    data1['initial'] = data1['author'].str[0]
    data2['initial'] = data2['author'].str[0]
    
    allinitials = set(data1.initial).intersection(set(data2.initial))
    matches = 0
    
    for initial in allinitials:  
        print(initial)
        df1 = data1.loc[data1.initial == initial, : ]
        df2 = data2.loc[data2.initial == initial, : ]
        print(len(df1['author']), len(df2))
        
        for idx, row in df1.iterrows():
            author = row.author
            if author != author: # do not count the case when author is nan
                continue
            title = row.title
            if pd.isnull(title):
                    continue
                    
            year = row.year
            
            idxlist = []
            
            for idx2, row2 in df2.iterrows():
                author2 = row2.author
                if author2 != author2: # do not count the case when author is nan
                    continue
               
                aRatio = fuzz.ratio(str(author), str(author2))
                if aRatio < auththreshold:
                    continue
                
                title2 = row2.title
                if pd.isnull(title2):
                    continue
                
                if title.startswith('the '):        # remove initial the before truncation
                    title = title[4:]               # this is commonly missing in the reviews
                if title2.startswith('the '):
                    title2 = title2[4:]
                
                if truncate == 0:
                    new_title = title
                    new_title2 = title2
                    
                else:       
                    shorter_title = min(len(title), len(title2), truncate) # determining the truncate length: shortest of two titles and truncate number
                    new_title = title[:shorter_title]
                    new_title2 = title2[:shorter_title]
                
                minlen = min(len(new_title), len(new_title2))
                
                if minlen < 3:
                    continue
                
                if minlen < 12:
                    handicap = 12 - minlen
                elif minlen > 24:
                    handicap = 24 - minlen  # this will actually give an advantage to long titles, up to truncate max
                else:
                    handicap = 0
                
                    
                year2 = row2.year
                    
                tRatio = fuzz.ratio(str(new_title), str(new_title2)) - handicap
                original_tRatio = fuzz.ratio(str(new_title), str(new_title2))
                
                totalmatch = tRatio + (aRatio / 1.5)
                
                if tRatio >= titlethreshold:
                    idxlist.append((totalmatch, idx2, aRatio, tRatio, original_tRatio, new_title2, year2, row2.rownum))
            
            if len(idxlist) > 0:
                idxlist.sort(reverse = True)
                row2 = df2.loc[idxlist[0][1], : ]
                aRatio = idxlist[0][2]
                tRatio = idxlist[0][3]
                original_tRatio = idxlist[0][4]
                new_title2 = idxlist[0][5]
                year2 = idxlist[0][6]
                rownum = idxlist[0][7]
                
                volume_author_list.append(author)
                author2 = row2.author
                review_author_list.append(author2)
                
                aRatio_list.append(aRatio)
                old_volume_title = row.title
                volume_title_list.append(old_volume_title)
                truncated_volume_title_list.append(new_title)
                old_review_title = row2.title
                review_title_list.append(old_review_title)
                truncated_rev_title_list.append(new_title2)
                
                tRatio_list.append(tRatio)
                original_tRatio_list.append(original_tRatio)
                volume_year_list.append(year)
                review_year_list.append(year2)
                docid_list.append(row.docid)
                rownum_list.append(rownum)
                
                matches += 1
                if matches % 100 == 1:
                    print('matches: ', matches)
    #print(volume_author_list, review_author_list, aRatio_list, truncated_volume_title_list, volume_title_list, truncated_rev_title_list, review_title_list,
    #                              tRatio_list, original_tRatio_list, volume_year_list, review_year_list, docid_list, rownum_list)                
    df_output = pd.DataFrame(list(zip(volume_author_list, review_author_list, aRatio_list, truncated_volume_title_list, volume_title_list, truncated_rev_title_list, review_title_list,
                                  tRatio_list, original_tRatio_list, volume_year_list, review_year_list, docid_list, rownum_list)), columns=['volume_author', 'review_author',
                                                                                              'aRatio', 'trunc_volumetitle', 'volume_title', 'trunc_revtitle', 'review_title',
                                                                                              'tRatio', 'original_tRaio', 'volume_year', 'review_year', 'docid', 'rownum'])
    return df_output
```

```python
fuzzywuzzy_clean = fuzzywuzzy_matching(volume_title_author_year, review_title_author_year, 30, 60, 60)
fuzzywuzzy_clean = fuzzywuzzy_clean.sort_values(by=['tRatio'], ascending=False).reset_index(drop=True)
fuzzywuzzy_clean
```
